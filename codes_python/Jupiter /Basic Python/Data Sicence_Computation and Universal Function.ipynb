# Numpy

```python
import numpy as np
np.random.seed(0)
```

```python
def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output
```

```python
big_array = np.random.randint(1,100,size=1000000)
```

```python
%timeit compute_reciprocals(big_array)
```

```python
%timeit (1.0 / big_array)
```

```python
x = np.arange(4)
print(x)
print(x/2)
print(x+1)
print(x**2)
print(abs(x-1))
```

```python
theta = np.linspace(0,np.pi,4)
print(theta)
print(np.sin(theta))
print(np.cos(theta))
```

```python
x = np.arange(1,4)
print(np.exp(x))
print(np.exp2(x))
print(np.power(3,x))
print(np.power(x,2))
print(np.log(x))
print(np.log2(x))
print(np.log10(x))
```

```python
x = [0,0.001,0.01,0.1]
print(np.expm1(x)) #e^x-1
print(np.log1p(x)) #log(1+x)
```

# Scipy

```python
from scipy import special
```

```python
x = [1,5,10]
print(special.gamma(x))
print(special.gammaln(x))
print(special.beta(x,2))
```

# Advanced ufuncs

```python
#指定输出位置
x = np.arange(5)
y = np.empty(5)
np.multiply(x,10,out=y)
print(y)
```

```python
y = np.zeros(10)
np.power(2,x,out=y[::2])
print(y)
```

```python
x = np.arange(1,5)
print(x)
print(np.add.reduce(x))
print(np.multiply.reduce(x))
print(np.add.accumulate(x))
print(np.multiply.accumulate(x))
print(np.add.outer(x,x))
print(np.multiply.outer(x,x))
```

# Aggregations

```python
L = np.random.random(100)
```

```python
print(sum(L))
print(np.sum(L))
```

```python
%timeit sum(L)
%timeit np.sum(L)
```

```python
print(np.min(L))
print(np.max(L))
```
