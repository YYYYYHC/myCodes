## 常用类型

## Numpy数组

```python
x = np.random.randint(10,size=(3,4,5))
print(x.ndim)
print(x.shape)
print(x.size)
print(x.dtype)
print(x.itemsize)
print(x.nbytes)
```

```python
# print(x)
x2 = x[0]
print(x2)
```

```python
print(x2[:,3])
print(x2[3,1:2])
print(x2[1::3,:])#从1开始，步长为3（三个一读，跳过两个）
```

```python
x3=x2[:2,:2]
print(x3)
x3[0,0]=1
print(x2)
```

```python
xc=x2[:2,:2].copy()
print(xc)
xc[0,0]=2
print(x2)
```

```python
grid = np.arange(1,10).reshape((3,3))
print(grid)
```

```python
print(np.concatenate([grid,grid]))
print(np.concatenate([grid,grid],axis=1))
```

```python
x = np.array([1,2,3])
grid = np.arange(9,3,-1).reshape(2,3)
print(x)
print(grid)
```

```python
print(np.vstack([x,grid]))
y=np.array([1,2]).reshape(2,1)
print(y)
print(np.hstack([grid,y]))
```

```python
x = np.arange(10)
print(x)
x1,x2,x3 = np.split(x,[3,6])
print(x1,x2,x3)
```

```python
x1,x2 = np.split(x,[9])
x1 = x1.reshape(3,3)
print(x1)
```

```python
upper,lower=np.vsplit(x1,[1])
print(upper)
print(lower)
left,right=np.hsplit(x1,[2])
print(left)
print(right)
```
